```python
import sqlite3
from pathlib import Path
db = str(Path("~/Documents/Timekeeper/timekeeper.db").expanduser())  # adjust if different
with sqlite3.connect(db) as conn:
    rows = conn.execute("SELECT substr(start_time,1,10), COUNT(*) FROM time_entries WHERE manually_added=0 GROUP BY substr(start_time,1,10) ORDER BY 1").fetchall()
for date, count in rows:
    print(date, count)
```
